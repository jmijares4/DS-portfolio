```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow import keras
```

```python
celsius = np.array([-40, -10, 0, 8, 15, 22, 38], dtype=float)
farenheit = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)
```

## Import dataframe so the AI can work with and learn.

```python
capa = tf.keras.layers.Dense(units=1, input_shape=[1])
modelo = tf.keras.Sequential([capa])
```

## Since is a farly simple model, we are using just 1 layer.

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

## The smaller the Adam, the better loss, but at the cost of more epochs needed.

## ItÂ´s better to have many epochs with minimal loss, than few epochs with big loss.

```python
historial = modelo.fit(celsius, farenheit, epochs=1000)
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoch")
plt.ylabel("Loss Magnitude")
plt.plot(historial.history["loss"])
```

## Looking at the "Loss plot" it can be seen that  the loss was near zero at the 600Â´s epochs, but after a visual inspection it can be seen that the model is capable of a loss below 0.1, wich is reached after near 900 epochs, so for this spescific case 1000 epochs was a good choice for a high prescition need.

```python
temperatura = 100.0
resultado = modelo.predict([temperatura])
print(str(temperatura) + "celsius degrees are the same as " + str(resultado) + " farenheit degrees")
```

## Correct answer is 212 vs the result of 211.74065, so the model is successfully trained.
